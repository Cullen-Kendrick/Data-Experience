# Exploration

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/Cullen-Kendrick/Data-219/main/clean_data.csv")
```

**Bar Plot of Happiness Across Regions**

```python
hap_region = df.groupby('Region')['happy'].mean()
hap_region.plot.bar()
```

*Bar Plot Review*
---
This bar plot shows that reported happiness across the different regions of the world are all around the same. It is close to a normal distribution however, the variance between each region is to great to be called an actual normal distribution. North America and ANZ has the highest mean reported happiness, and South Asia has the lowest mean reported happiness.

# Scatter Plots

*GDP on Happiness*
---
In the general scatter plot, the relationship between GDP and happiness seems to be very solid. The region with the r squared would probably be Western Europe, and the highest is probably the Commonwealth of Independent States or South Asia, were there seems to be no relationship.

**General Scatter Plot for Graphing Happiness on GDP**

```python
# Scatter plot of Happy by:
    # log GDP per capita
sns.scatterplot(data=df, x='GDP', y='happy')
```

**Region Specific GPD-Happy Scatter Plot**

```python
sns.relplot(data=df, x='GDP', y='happy', col='Region', col_wrap=5)
```
